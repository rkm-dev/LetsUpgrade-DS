### Task 22 : Take a sample of size 50 from the column damageDealt for 100 times and calculate its mean. Plot it on a histogram and comment on its distribution.

```python
mean_lst = []
for i in range(100):
    mean_lst.append(df1.damageDealt.sample(50).mean())
    
    
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
axe = axe.flatten() 
    
sns.histplot(mean_lst, ax=axe[0], kde=True);
sns.histplot(pd.DataFrame(mean_lst, columns=["mean"]), ax=axe[1], palette="Set2", kde=True);
```
